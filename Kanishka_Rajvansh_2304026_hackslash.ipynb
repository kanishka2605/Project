```python
def fraction_missing(df):
    missed_frac=df.isnull().sum()/len(df)
    missed_frac=missed_frac[missed_frac>0].sort_values
    return missed_frac
```

```python
#Fraction of values missing
fraction_missing(df)
```

```python
def column_types(Dataframe):
    numerical=df.select_dtypes(include='number').columns
    categorical = df.select_dtypes(exclude=["number","bool_","object_"])
    object = df.select_dtypes(include='object').columns
    total_columns=len(df.columns)
    numerical_fraction = len(numerical)/total_columns
    categorical_fraction = len(categorical)/total_columns
    text_fraction = len(object)/total_columns
    return {
        'numerical':numerical_fraction,
        'categorical':categorical_fraction,
        'text':text_fraction
    }
```

```python
#Fraction of numerical,categorical,text variables
column_types(df)
```

```python
#Handling Missing Data
```

```python
df[df['Gender'].isnull()].index
```

```python
df=df.drop(index=[16, 304, 321, 361, 370],axis=0)
```

```python
df[df['Dependents'].isnull()].index
```

```python
df=df.drop(index=[140, 185, 190, 212, 213, 222, 327, 375],axis=0)
```

```python
num = df['Self_Employed'].mode()[0]
df['Self_Employed'].fillna(num, inplace=True)
```

```python
df['Credit_History']=df['Credit_History'].fillna(0)
```

```python
df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(0)
```

```python
#Doing some feature engineering
```

```python
df=df.drop('Loan_ID',axis=1)
```

```python
#Logistic Regression Model
```

```python
X=df.drop('Loan_Status',axis=1)
```

```python
y=df['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
```

```python
numerical_features = X.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X.select_dtypes(include=['object']).columns
```

```python
categorical_features
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X[numerical_features] = scaler.fit_transform(X[numerical_features])
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model=LogisticRegression()
```

```python
log_model.fit(X_train,y_train)
```

```python
y_pred=log_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print (classification_report(y_test,y_pred))
```

```python
#KNN Model
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
```

```python
operations=[('scaler',scaler),('knn',knn)]
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe=Pipeline(operations)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
k_values=list(range(1,20))
param_grid={'knn__n_neighbors':k_values}
full_cv_classifier=GridSearchCV(pipe,param_grid,cv=5,scoring='accuracy')
```

```python
full_cv_classifier.fit(X_train,y_train)
```

```python
full_pred=full_cv_classifier.predict(X_test)
```

```python
#Decision Tree model
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier() 
```
